```python
my_list = []
for x in range(1, 20):
    my_list.append(x)
print(my_list)
```

```python
print(my_list[0])
mid = (0 + len(my_list))//2
print(my_list[mid])
print([len(my_list)])
```

```python
print(my_list[0:5])
print(my_list[mid:]) #mid to last
print(my_list[-5:])
```

```python
sqr = [x**2 for x in range(1,10) if x%2==0]
print(sqr)
```

```python
import random

random_numbers = [random.randint(1, 20) for _ in range(15)]
print(f"Original list: {random_numbers}")

sorted_numbers = sorted(random_numbers)
print(f"ascending order {sorted_numbers}")

sorted_numbers = sorted(random_numbers, reverse= True)
print(f"ascending order {sorted_numbers}")

unique_number = sorted(set(random_numbers))
print(f"unique only in list {unique_number}")
```

```python
def reverse_list(my_list):
    return my_list[::-1]
original_lst = [1,2,3,4,5,6]
reversed_list = reverse_list(original_lst);
print(f"original list{original_lst}")
print(f"reverse list{reversed_list}")
```
